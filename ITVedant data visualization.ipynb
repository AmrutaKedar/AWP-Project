```python
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.stackplot(days,sleeping,eating, working ,playing, colors=['m','c','r','k'],labels=['sleeping','eating', 'working','playing'])

plt.xlabel('days')
plt.ylabel('total hrs spend')
plt.title('stack of day time utilize')
plt.legend()
plt.show()
```

```python
hours=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','g']
```

```python
plt.pie(hours,labels=activities,colors=cols)
plt.title("monday time utilization")
plt.legend()
plt.show()
```

```python
plt.pie(hours,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%')
```
